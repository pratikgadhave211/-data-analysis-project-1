```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


#LOAD AND INSPECT THE DATASET: 


data=pd.read_csv('file (4).csv')
# inspect the dataset
print(data.head())
# basic info about data
print("shape of dataset: ",data.shape)
print("column names: ",data.columns)

# DATA CLEANING:


# cleaning and removing duplicate files from data
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)

# EXPLODATORY DATA ANALYSIS(EDA):


# convert date column to date time format
data['Release_Date']=pd.to_datetime(data['Release_Date'],format='mixed')

# show about how much movie and tv shows added per year
show_added=data['Release_Date'].dt.year.value_counts().sort_index()
print("the data of show added per year is:  ",show_added)
# a line graph showing show added per year data
plt.figure(figsize=(10,5))
plt.plot(show_added.index,show_added.values,linestyle='--',marker='o',color='green',label='show added per year')
plt.grid(True,linestyle=':')
plt.xlabel('year')
plt.ylabel('no. of shows added per year')
plt.title('show added per year line Graph')
plt.legend()
plt.tight_layout()
plt.show()



# which year most show were added
print('The year  when  most content were added on netflix added is : ',show_added.idxmax(),'which were ',show_added.max())

# which year least content were added
print('The year  when  least content were added on netflix added is : ',show_added.idxmin(),'which were ',show_added.min()) 


# movie vs tv show graph

counts=data['Category'].value_counts()
plt.figure(figsize=(10,5))
plt.bar(counts.index,counts.values,color=['orange','red'],label=['movies','tv shows'])
plt.xlabel('content')
plt.ylabel('no.of shows')
plt.title('comparing tv shows and movies')
plt.tight_layout()
plt.legend()
plt.show()
# to no. of tv shows and movies present
print('total no. of movies are ',counts['Movie'])
# total no. of tv shows
print('total no. of tv shows are : ',counts['TV Show'])
# top 10 countries by content creation
top_10_countries=data['Country'].value_counts().head(10)
print('top ten countries with content creation are ',top_10_countries.index)
plt.figure(figsize=(10,5))
plt.bar(top_10_countries.index,top_10_countries.values,color='red',label='total content created by a country')
plt.xlabel('country')
plt.ylabel('no. of content created')
plt.legend()
plt.title('top ten contries in content creation')
plt.tight_layout()
plt.show()
# top 10 genre in the contents on netflix
top_10_genre=data['Type'].value_counts().head(10)
plt.figure(figsize=(10,5))
plt.pie(top_10_genre.values,labels=top_10_genre.index,autopct='%1.1f%%')
plt.tight_layout()
plt.show()
    
#  SPECIFIC INSIGHTS

# movies featuring tom cruise
tom_cruise_movies= data[data['Cast'].str.contains('Tom Cruise')]
print(' Tom Cruise featured in movies: ')
print(tom_cruise_movies[['Title','Type','Release_Date']])


print('this was a short data analysis project on netflix dataset!!')
```
